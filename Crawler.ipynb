# 目前只使用以下四個package 抓取網頁資訊

```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'http://www.stockq.org/'
req = requests.get(url)
if(req.status_code == 200):
    soup = BeautifulSoup(req.text.replace('\n',''),'lxml')
    print(soup)
```

## 目前只抓取index

```python
indexData=[]
trs = soup.find_all('tr' , {'class' ,re.compile('^row[1-6]')})
for tr in trs:
    for title in tr.find_all('a'):
        index = re.split('^/index/+',title['href'])
        if(len(index)>1):
            indexData.append(['StockQ',title.text,index[1].replace('.php','')])
        else:
            continue
indexData
```

```python
index_df = pd.DataFrame(indexData,columns=["From","Name","Key"])
index_df
```

使用已經抓取到的key定義function 取得價格資訊
--------------------------------------------------------------------

```python
def findPrice(Key):
    tr = soup.find('a',{'href':re.compile(Key)}).find_parent('tr') 
    print([td.text for td in tr.find_all('td')])  
```

```python
findPrice('TPX')
```
